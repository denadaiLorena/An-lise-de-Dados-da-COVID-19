# Importação das Bibliotecas
Importar a biblioteca pandas para manipulação e análise de dados estruturados.

```python
import pandas as pd
```

# Carregamento do DataFrame
Carregar o conjunto de dados Parquet filtrado para análise.

```python
df = pd.read_parquet('dados_es_filtrados.parquet')
df.head()
```

# Filtragem das Colunas de Comorbidade
Isolar colunas que contenham o termo 'Comorbidade'.

```python
comorb_cols = [col for col in df.columns if 'Comorbidade' in col]
print(comorb_cols)
```

# Inspeção dos Tipos de Dados (dtypes)
Visualizar os tipos das colunas filtradas.

```python
df[comorb_cols].dtypes
```

# Contagem de Valores 'Ignorado'
Contar quantas vezes o termo 'Ignorado' aparece em cada coluna de comorbidade.

```python
# Contagem exata do termo 'Ignorado' por coluna
ignorado_counts = (df[comorb_cols] == 'Ignorado').sum()
print("Quantidade de 'Ignorado' por coluna:")
print(ignorado_counts)

# Porcentagem em relação ao total
total_rows = len(df)
print("\nPorcentagem de 'Ignorado' por coluna:")
print((ignorado_counts / total_rows) * 100)
```

```python
# Verificar os valores únicos presentes em uma das colunas para entender como os dados estão vindo
print("Valores únicos na coluna ComorbidadePulmao:")
print(df['ComorbidadePulmao'].unique())

# Verificar se existem valores nulos ou vazios
print("\nValores nulos/vazios por coluna:")
print(df[comorb_cols].isna().sum())
```
